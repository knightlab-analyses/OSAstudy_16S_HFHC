```python
# first we import the metdata into pandas
mf = pd.read_csv('../data/Microbiome/longIHC_metadata_cleaned.txt', sep='\t',index_col=0)
# next we aggregate by subjects (i.e. 'host_subject_id') 
mf = mf.groupby('host_subject_id').agg({'diet_exp':'first','diet':'first', 'exposure_type':'first', 'cage_number':'first'})
# now we save the metadata in QIIME2 format.
mf.index.name = '#SampleID'
mf.index = mf.index.astype(str)
mf.to_csv('../data/Microbiome/lr-metrics/tempted/tempted-metadata.tsv', sep='\t')
mf.head()
```

```python
# import log-ratio data
metadata_one = pd.read_csv('../data/Microbiome/longIHC_metadata_cleaned.txt',
                           sep='\t', index_col=0)
# import rest of the metadata
metadata_two = pd.read_csv('../data/Microbiome/lr-metrics/tempted/longIHC_tempted_topbot10num_Axis1_lr.tsv',
                           sep='\t', index_col=0)[['Current_Natural_Log_Ratio']]
#rename LR column so you know what they are
metadata_two.rename(columns={'Current_Natural_Log_Ratio':'TopBot10Num_Axis1'}, inplace=True)
# merge the data
log_ratio_metdata = pd.concat([metadata_two, metadata_one], axis=1)
#drop rows with no value for LR
log_ratio_metdata = log_ratio_metdata.dropna(subset=['TopBot10Num_Axis1'])
# export in QIIME2 format
log_ratio_metdata.index.name = '#SampleID'
log_ratio_metdata.to_csv('../data/Microbiome/lr-metrics/tempted/lr_merged-data.tsv', sep='\t')
log_ratio_metdata.head()
```

```python
# import log-ratio data
metadata_one = pd.read_csv('../data/Microbiome/longIHC_metadata_cleaned.txt',
                           sep='\t', index_col=0)
# import rest of the metadata
metadata_two = pd.read_csv('../data/Microbiome/lr-metrics/tempted/tempted-qurro_Muribaculaceae-Akkermansiaceae_LR.tsv',
                           sep='\t', index_col=0)[['Current_Natural_Log_Ratio']]
#rename LR column so you know what they are
metadata_two.rename(columns={'Current_Natural_Log_Ratio':'Muribaculaceae_Akkermansiaceae'}, inplace=True)
# merge the data
log_ratio_metdata = pd.concat([metadata_two, metadata_one], axis=1)
#drop rows with no value for LR
log_ratio_metdata = log_ratio_metdata.dropna(subset=['Muribaculaceae_Akkermansiaceae'])
# export in QIIME2 format
log_ratio_metdata.index.name = '#SampleID'
log_ratio_metdata.to_csv('../data/Microbiome/lr-metrics/tempted/MuribaculaceaeAkkermansiaceae_LR_merged.tsv', sep='\t')
log_ratio_metdata.head()
```
